#### 1. Import pandas library

```python
import pandas as pd 
```

#### 2. Import users table:

```python
df_user = pd.read_csv("users_table.csv")
df_user.head()
```

#### 3. Rename Id column to userId

```python
df_user.rename(columns = {'Id':'userId'}, inplace = True)
df_user.head()
```

#### 4. Import posts table:

```python
df_post = pd.read_csv("posts_table.csv")
df_post.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
df_post.rename(columns = {'Id':'postId','OwnerUserId':'userId'}, inplace = True)
df_post.head()
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
df_user_2 = df_user[['userId', 'Reputation','Views','UpVotes','DownVotes']]
df_post_2 = df_post[['postId', 'Score','userId','ViewCount','CommentCount']]
df_user_2.head()
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
df_merge = df_user_2.merge(df_post_2, left_on='userId', right_on='postId')
df_merge.head()
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
count_nan = df_merge.isna().sum().sum()
count_nan
```

```python
df_merge.info()
```

I´m going to clean the 'userId' because filling with a value could compute next columns into wrong users or non existing users.
